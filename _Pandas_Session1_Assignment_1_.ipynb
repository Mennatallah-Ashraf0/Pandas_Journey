### Task 5.1: Using map() (5 points)

1. Create 'Attrition_Binary': map 'Yes' to 1, 'No' to 0
2. Create 'Gender_Full': map 'Male' to 'Male Employee', 'Female' to 'Female Employee'
3. Create 'Satisfaction_Label': map 1='Low', 2='Medium', 3='High', 4='Very High'

```python
###1 Create 'Attrition_Binary': map 'Yes' to 1, 'No' to 0
attrition_dic = {'Yes' : 1, 'No' : 0}
df['Attrition_Binary'] = df['Attrition'].map(attrition_dic)
df['Attrition_Binary']
```

```python
df['Gender'].unique()
```

```python
###2 Create 'Gender_Full': map 'Male' to 'Male Employee', 'Female' to 'Female Employee'
df['Gender'] = df['Gender'].astype(str).str.strip()
gender_dic = {'Male' : 'Male Employee', 'Female' : 'Female Employee'}
df['Gender_Full'] = df['Gender'].map(gender_dic)
df['Gender_Full']
```

```python
###3 Create 'Satisfaction_Label': map 1='Low', 2='Medium', 3='High', 4='Very High'
satisfaction_dic = {1 : 'Low', 2 : 'Medium', 3 : 'High', 4 : 'Very High'}
df['Satisfaction_Label'] = df['JobSatisfaction'].map(satisfaction_dic)
df['Satisfaction_Label']
```

### Task 5.2: Using apply() with Lambda (10 points)

1. Create 'AgeGroup':
   - Under 30: 'Young'
   - 30-45: 'Mid-Career'
   - Over 45: 'Senior'

2. Create 'IncomeLevel':
   - Below 3000: 'Low'
   - 3000-7000: 'Medium'
   - Above 7000: 'High'

3. Create 'TenureRatio': YearsAtCompany / TotalWorkingYears (handle division by zero)

```python
###1 Create 'AgeGroup': Under 30: 'Young' 30-45: 'Mid-Career' Over 45: 'Senior'
df['AgeGroup'] = df['Age'].apply(lambda x :'Young' if x < 30 else ('Mid-Career'if x <= 45 else 'Senior' ))
df['AgeGroup']
```

```python
###2 Create 'IncomeLevel': Below 3000: 'Low' 3000-7000: 'Medium' Above 7000: 'High'
df['IncomeLevel'] = df['MonthlyIncome'].apply(lambda x : 'low' if x < 3000 else('medium' if x<= 7000 else 'high'))
df['IncomeLevel']
```

```python
###3 Create 'TenureRatio': YearsAtCompany / TotalWorkingYears (handle division by zero)
df['TenureRatio'] = df.apply(lambda row : row['YearsAtCompany'] / row['TotalWorkingYears'] if row['TotalWorkingYears'] >0 else 0, axis = 1 )
df['TenureRatio']
```

### Task 5.3: Custom Functions with apply() (5 points)

1. Write a function that returns 'High Risk' if JobSatisfaction <= 2 AND WorkLifeBalance <= 2, else 'Normal'
   Apply to create 'RiskCategory' column

2. Calculate 'AnnualIncome' (MonthlyIncome * 12) using apply

```python
###1 Write a function that returns 'High Risk' if JobSatisfaction <= 2 AND WorkLifeBalance <= 2, else 'Normal' Apply to create 'RiskCategory' column
def calculate_risk(row):
    if row['JobSatisfaction'] <= 2 and row['WorkLifeBalance'] <= 2:
        return 'High Risk'
    else:
        return 'Normal'

df['RiskCategory'] = df.apply(calculate_risk, axis=1)
df['RiskCategory']
```

```python
###2 Calculate 'AnnualIncome' (MonthlyIncome * 12) using apply
df['AnnualIncome'] = df['MonthlyIncome'].apply(lambda x: x * 12)
df['AnnualIncome']
```
