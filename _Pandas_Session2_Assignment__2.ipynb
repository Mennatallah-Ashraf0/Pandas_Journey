1. Create 'lead_time_category' using cut():

   - 8-30 days: 'Short Notice'
   - 31-90 days: 'Planned'
   - 90+ days: 'Far Advance'

### Task 4.4: Outlier Detection (5 points)

The avg_daily_rate column may have outliers.

1. Calculate the IQR (Interquartile Range)
2. Find outliers using: lower < Q1 - 1.5*IQR or upper > Q3 + 1.5*IQR
3. How many outliers exist?
4. Cap outliers at the upper/lower bounds (do not remove them)

```python
#1 Calculate the IQR (Interquartile Range)

Q1 = Hotel_Booking_Demand['avg_daily_rate'].quantile(0.25)
Q3 = Hotel_Booking_Demand['avg_daily_rate'].quantile(0.75)

IQR = Q3 - Q1
IQR
```

```python
#2 Find outliers using: lower < Q1 - 1.5IQR or upper > Q3 + 1.5IQR

lower_bound = Q1 - 1.5 * IQR

upper_bound = Q3 + 1.5 * IQR

print(lower_bound)
print(upper_bound)
```

```python
#3 How many outliers exist?

outliers_filter = ((Hotel_Booking_Demand['avg_daily_rate'] < lower_bound) | (Hotel_Booking_Demand['avg_daily_rate'] > upper_bound))

num_outliers = outliers_filter.sum()

num_outliers
```

```python
#4 Cap outliers at the upper/lower bounds (do not remove them)

Hotel_Booking_Demand['avg_daily_rate'] = Hotel_Booking_Demand['avg_daily_rate'].clip(lower=lower_bound, upper=upper_bound)
```
